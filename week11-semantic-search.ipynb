## *Task 1.2: Calculate Similarity* 

```python
import numpy as np

def cosine_similarity(vec1, vec2):

    vec1 = np.array(vec1)
    vec2 = np.array(vec2)

    dot_product = np.dot(vec1, vec2)

    norm1 = np.linalg.norm(vec1)
    norm2 = np.linalg.norm(vec2)

    return dot_product / (norm1 * norm2)


# Test phrases
phrases = [
    'vacation policy',
    'time off rules',
    'PTO guidelines',
    'dress code requirements'
]

# Get embeddings
embeddings = [get_embedding(p) for p in phrases]

# Compare first phrase with others
base = embeddings[0]

print(f'Comparing "{phrases[0]}" with:\n')

for i, phrase in enumerate(phrases[1:], 1):

    similarity = cosine_similarity(
        base,
        embeddings[i]
    )

    print(
        f'{phrase:30} Similarity: {similarity:.4f}'
    )
```

```python
# Install required packages (run once)
!pip install -q chromadb langchain-community langchain-text-splitters google-genai

# Imports
import chromadb

from google import genai

from kaggle_secrets import UserSecretsClient

from langchain_community.document_loaders import (
    DirectoryLoader,
    TextLoader
)

from langchain_text_splitters import (
    RecursiveCharacterTextSplitter
)

# =========================
# Load Gemini API Key
# =========================

user_secrets = UserSecretsClient()

api_key = user_secrets.get_secret("Gemini_API_Key")

# =========================
# Gemini Client
# =========================

client = genai.Client(api_key=api_key)

# =========================
# Embedding Function
# =========================

def get_embedding(text):

    response = client.models.embed_content(
        model="gemini-embedding-001",
        contents=text
    )

    return response.embeddings[0].values

# =========================
# ChromaDB Client
# =========================

chroma_client = chromadb.PersistentClient(
    path="./chroma_db"
)

# Create/Get Collection
collection = chroma_client.get_or_create_collection(
    name="company_documents"
)

# =========================
# Load Documents
# =========================

loader = DirectoryLoader(
    "/kaggle/input/datasets/rubabq66/company-documents",
    glob="*.txt",
    loader_cls=TextLoader
)

documents = loader.load()

print(f"Loaded {len(documents)} documents")

# =========================
# Split Documents into Chunks
# =========================

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = text_splitter.split_documents(documents)

print(f"Created {len(chunks)} chunks")

# =========================
# Add to ChromaDB
# =========================

if collection.count() == 0:

    chunk_texts = [
        chunk.page_content
        for chunk in chunks
    ]

    # Generate embeddings
    embeddings = [
        get_embedding(text)
        for text in chunk_texts
    ]

    # Store in ChromaDB
    collection.add(

        documents=chunk_texts,

        embeddings=embeddings,

        ids=[
            f"doc_{i}"
            for i in range(len(chunks))
        ],

        metadatas=[
            {
                "source": chunk.metadata.get(
                    "source",
                    "unknown"
                )
            }
            for chunk in chunks
        ]
    )

    print(f"Added {len(chunks)} chunks to ChromaDB")

else:

    print(
        f"Collection already contains "
        f"{collection.count()} chunks"
    )

# =========================
# Test Semantic Search
# =========================

query = "What is the vacation policy?"

query_embedding = get_embedding(query)

results = collection.query(
    query_embeddings=[query_embedding],
    n_results=2
)

print("\nTop Results:\n")

for i, doc in enumerate(results["documents"][0]):

    print(f"Result {i+1}:")
    print(doc)
    print("-" * 50)
```

```python
from kaggle_secrets import UserSecretsClient
from langchain_google_genai import ChatGoogleGenerativeAI

# =====================
# API KEY (Kaggle safe)
# =====================
user_secrets = UserSecretsClient()
api_key = user_secrets.get_secret("Gemini_API_Key")

# =====================
# LLM (Gemini)
# =====================
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0,
    google_api_key=api_key
)

# =====================
# RAG FUNCTION
# =====================
def semantic_rag(query, n_results=3):

    # Step 1: Vector search (IMPORTANT FIX)
    query_embedding = get_embedding(query)

    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=n_results
    )

    if not results["documents"][0]:
        return "I don't know based on the provided documents."

    # Step 2: Context
    context = "\n\n---\n\n".join(results["documents"][0])

    # Step 3: Prompt
    prompt = f"""
You are a helpful HR assistant.

Answer ONLY using the context below.

If the answer is not in the context, say:
"I don't know based on the provided documents."

Context:
{context}

Question:
{query}

Answer:
"""

    # Step 4: Generate answer
    response = llm.invoke(prompt)

    return response.content


# =====================
# TEST
# =====================
questions = [
    "How much time off do employees get?",
    "Can I work from home?",
    "What is the maternity leave policy?"
]

for q in questions:
    print("\nQ:", q)
    print("A:", semantic_rag(q))
```

## KEYWORD VS SEMANTIC COMPARISON*

```python
def keyword_search(query, chunks, top_k=3):

    query_lower = query.lower()

    scored = []

    for chunk in chunks:

        content_lower = chunk.page_content.lower()

        score = sum(
            content_lower.count(word)
            for word in query_lower.split()
        )

        if score > 0:

            scored.append((score, chunk))

    # Sort highest score first
    scored.sort(
        reverse=True,
        key=lambda x: x[0]
    )

    # Return only chunks
    return [
        c
        for s, c in scored[:top_k]
    ]


# Test on synonym query
query = 'PTO policy'

# Keyword search
kw_results = keyword_search(
    query,
    chunks,
    top_k=2
)

print('\nKEYWORD SEARCH:')

print(f'Found {len(kw_results)} results')

if kw_results:
    print(
        f'Top result: '
        f'{kw_results[0].page_content[:100]}...'
    )


# Semantic search
print('\nSEMANTIC SEARCH:')

sem_results = vector_search(
    query,
    n_results=2
)

print(
    f'Found '
    f'{len(sem_results["documents"][0])} '
    f'results'
)

print(
    f'Top result: '
    f'{sem_results["documents"][0][0][:100]}...'
)
```

The goal of this lab was to build a **Retrieval-Augmented Generation (RAG) system** and compare:

I also integrated **Google Gemini embeddings + LLM** with **ChromaDB vector database** to build a working HR document assistant.

#  3. Keyword-Based Search (Baseline Method)

##  Approach
I implemented a simple keyword matching system:

- Convert query to lowercase
- Count word matches in document chunks
- Rank based on frequency

##  Limitations

- No understanding of meaning
- Cannot detect synonyms
  - Example: "WFH" vs "remote work"
- Fails on paraphrased queries

---

#  4. Semantic Search (Embedding-Based)

##  Approach

I replaced keyword matching with embeddings:

- Used **Gemini embedding model**
- Converted text → high-dimensional vector
- Stored vectors in **ChromaDB**
- Compared similarity using distance metrics
